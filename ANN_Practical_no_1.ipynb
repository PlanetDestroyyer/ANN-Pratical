```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return (1/1+ np.exp(-x))
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
def linear(x):
  return x
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y_sigmoid = sigmoid(x)
```

```python
y_tanh = tanh(x)
```

```python
y_relu = relu(x)
```

```python
y_linear = linear(x)
```

```python
plt.figure(figsize=(10,6))
plt.title('activation functions')
plt.subplot(2,3,1)
plt.plot(x, y_sigmoid, label='sigmoid')
plt.grid()
plt.subplot(2,3,2)
plt.plot(x, y_tanh, label='tanh')
plt.grid()
plt.subplot(2,3,3)
plt.plot(x, y_relu, label='relu')
plt.grid()
plt.subplot(2,3,4)
plt.plot(x, y_linear, label='linear')
plt.grid()
```
